```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.scrapethissite.com/pages/forms/"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
soup.find('p').text
```

```python
table_data = soup.find_all('td')
```

```python
 table_heads = soup.find_all('th')
```

```python
 table = soup.find_all('tr')
```

```python
team_table_heads = [heads.text.strip() for heads in table_heads]
print(team_table_heads)
```

```python
import pandas as pd
```

```python
 df = pd.DataFrame(columns = team_table_heads)
```

```python
df
```

```python
column_data = table
```

```python
all_rows = []
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    all_rows.append(individual_row_data)
    print(individual_row_data)
df = pd.DataFrame(all_rows, columns = team_table_heads)
```
